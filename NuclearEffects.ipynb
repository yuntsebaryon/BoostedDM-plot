```python
import ROOT
import math
ROOT.gStyle.SetOptStat(0)
f1 = ROOT.TFile('/Users/yuntse/Data/pheno/BoostedDM/dune_v07_06_01/bdm/ana/dune_fermion_e20_m10_g1_z1_fixedflux.0_Gen_g4_RecoSmear_ana.root')
f2 = ROOT.TFile('/Users/yuntse/Data/pheno/BoostedDM/dune_v07_06_01/bdm/ana/dune_fermion_e20_m10_g1_z1_nonuclear.0_Gen_g4_ana.root')
```

```python
t1 = f1.Get('MCParticles')
t2 = f2.Get('MCParticles')
```

```python
h1VisibleAngle = ROOT.TH1D('NEVisibleAngle', 'Visible angle with nuclear effects', 50, -1., 1.)
h2VisibleAngle = ROOT.TH1D('VisibleAngle', 'Visible angle without nuclear effects', 50, -1., 1.)
h1VisibleAngleNoN = ROOT.TH1D('NEVisibleAngleNoN', 'Visible angle with nuclear effects but no neutron', 50, -1., 1.)
h2VisibleAngleNoN = ROOT.TH1D('VisibleAngleNoN', 'Visible angle without nuclear effects, no neutron', 50, -1., 1.)
h1LeadingAngle = ROOT.TH1D('NELeadingAngle', 'Leading angle with nuclear effects', 50, -1., 1.)
h2LeadingAngle = ROOT.TH1D('LeadingAngle', 'Leading angle without nuclear effects', 50, -1., 1.)
```

```python
for i in t1:
    VisibleCosAngle = math.cos(i.VisibleAngle[0])
    h1VisibleAngle.Fill(VisibleCosAngle)
    VisibleCosAngleNoN = math.cos(i.VisibleNoNAngle[0])
    h1VisibleAngleNoN.Fill(VisibleCosAngleNoN)
    LeadingCosAngle = math.cos(i.LeadingParticleAngle[0])
    h1LeadingAngle.Fill(LeadingCosAngle)
```

```python
for i in t2:
    VisibleCosAngle = math.cos(i.VisibleAngle[0])
    h2VisibleAngle.Fill(VisibleCosAngle)
    VisibleCosAngleNoN = math.cos(i.VisibleNoNAngle[0])
    h2VisibleAngleNoN.Fill(VisibleCosAngleNoN)
    LeadingCosAngle = math.cos(i.LeadingParticleAngle[0])
    h2LeadingAngle.Fill(LeadingCosAngle)
```

```python
cVisibleAngle = ROOT.TCanvas("VisibleAngle", "Visible CosAngle", 1600, 1200)
cVisibleAngle.SetBottomMargin(0.15)
h2VisibleAngle.SetLineColor(906)
h2VisibleAngle.SetLineWidth(3)
h2VisibleAngle.SetTitle("")
h2VisibleAngle.GetXaxis().SetTitle("cos(#theta)")
h2VisibleAngle.GetXaxis().SetTitleSize(0.07)
h2VisibleAngle.GetXaxis().SetLabelSize(0.06)
h2VisibleAngle.GetYaxis().SetLabelSize(0.06)
h2VisibleAngle.Draw()
h1VisibleAngle.SetLineColor(593)
h1VisibleAngle.SetLineWidth(3)
h1VisibleAngle.Draw("same")
lVisibleAngle = ROOT.TLegend( 0.15, 0.6, 0.75, 0.8 )
lVisibleAngle.SetBorderSize(0)
lVisibleAngle.SetFillStyle(0)
lVisibleAngle.AddEntry( h1VisibleAngle, "With Nuclear Effects")
lVisibleAngle.AddEntry( h2VisibleAngle, "No Nuclear Effects")
lVisibleAngle.Draw()
cVisibleAngle.Draw()
cVisibleAngle.SaveAs('VisibleAngle.png')
```

```python
cVisibleAngleNoN = ROOT.TCanvas("VisibleNoNAngle", "Visible CosAngle without neutrons", 1600, 1200)
cVisibleAngleNoN.SetBottomMargin(0.15)
h2VisibleAngleNoN.SetLineColor(906)
h2VisibleAngleNoN.SetLineWidth(3)
h2VisibleAngleNoN.SetTitle("")
h2VisibleAngleNoN.GetXaxis().SetTitle("cos(#theta)")
h2VisibleAngleNoN.GetXaxis().SetTitleSize(0.07)
h2VisibleAngleNoN.GetXaxis().SetLabelSize(0.06)
h2VisibleAngleNoN.GetYaxis().SetLabelSize(0.06)
h2VisibleAngleNoN.Draw()
h1VisibleAngleNoN.SetLineColor(593)
h1VisibleAngleNoN.SetLineWidth(3)
h1VisibleAngleNoN.Draw("same")
lVisibleAngleNoN = ROOT.TLegend( 0.15, 0.6, 0.75, 0.8 )
lVisibleAngleNoN.SetBorderSize(0)
lVisibleAngleNoN.SetFillStyle(0)
lVisibleAngleNoN.AddEntry( h1VisibleAngle, "With Nuclear Effects")
lVisibleAngleNoN.AddEntry( h2VisibleAngle, "No Nuclear Effects")
lVisibleAngleNoN.Draw()
cVisibleAngleNoN.Draw()
cVisibleAngleNoN.SaveAs('VisibleNoNAngle.png')
```

```python
cLeadingAngle = ROOT.TCanvas("LeadingAngle", "Leading CosAngle", 1600, 1200)
cLeadingAngle.SetBottomMargin(0.15)
h2LeadingAngle.SetLineColor(906)
h2LeadingAngle.SetLineWidth(3)
h2LeadingAngle.SetTitle("")
h2LeadingAngle.GetXaxis().SetTitle("cos(#theta)")
h2LeadingAngle.GetXaxis().SetTitleSize(0.07)
h2LeadingAngle.GetXaxis().SetLabelSize(0.06)
h2LeadingAngle.GetYaxis().SetLabelSize(0.06)
h2LeadingAngle.Draw()
h1LeadingAngle.SetLineColor(593)
h1LeadingAngle.SetLineWidth(3)
h1LeadingAngle.Draw("same")
lLeadingAngle = ROOT.TLegend( 0.15, 0.6, 0.75, 0.8 )
lLeadingAngle.SetBorderSize(0)
lLeadingAngle.SetFillStyle(0)
lLeadingAngle.AddEntry( h1VisibleAngle, "With Nuclear Effects")
lLeadingAngle.AddEntry( h2VisibleAngle, "No Nuclear Effects")
lLeadingAngle.Draw()
cLeadingAngle.Draw()
cLeadingAngle.SaveAs('LeadingAngle.png')
```
