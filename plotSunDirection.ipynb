```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inFile = '/Users/yuntse/data/pheno/BoostedDM/dune_v07_06_01/atmos/background_sun_pos.dat'
```

```python
SunDirection = np.loadtxt( inFile )
```

```python
SunDirectionPolar = np.empty( [len(SunDirection), 3] )
```

```python
for i in range( len(SunDirection) ):
    SunDirectionPolar[i,0] = np.sqrt(SunDirection[i,0]**2+SunDirection[i,1]**2+SunDirection[i,2]**2)
    SunDirectionPolar[i,1] = SunDirection[i,2]/SunDirectionPolar[i,0]
    SunDirectionPolar[i,2] = np.arctan2(SunDirection[i,1], SunDirection[i,0])
```

```python
plt.hist2d( SunDirectionPolar[:,1], SunDirectionPolar[:,2], 100)
plt.xlabel(r'cos$\theta$')
plt.ylabel(r'$\phi$')
plt.savefig('SunDirection.pdf')
```
