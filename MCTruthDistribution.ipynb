```python
import ROOT
f = ROOT.TFile("/Users/yuntse/Work/pheno/BoostedDM/gallery/srcs/MCDistributions.root", "READ")
ROOT.gStyle.SetOptStat(0)
```

```python
particles = ['Proton', 'Neutron', 'Pion', 'DM', 'Other']

hN = {}
cN = {}
for particle in particles:
    hN[particle] = f.Get("n%ss" % particle )
    hN[particle].SetLineColor(591)
    hN[particle].SetLineWidth(3)
    hN[particle].SetTitle("")
    hN[particle].GetXaxis().SetTitle("Number of %ss" % particle)
    hN[particle].GetXaxis().SetTitleSize(0.06)
    hN[particle].GetXaxis().SetTitleFont(132)
    hN[particle].GetXaxis().SetLabelSize(0.06)
    hN[particle].GetXaxis().SetLabelFont(132)
    hN[particle].GetYaxis().SetTitle("Entries")
    hN[particle].GetYaxis().SetTitleSize(0.07)
    hN[particle].GetYaxis().SetTitleFont(132)
    hN[particle].GetYaxis().SetLabelSize(0.06)
    hN[particle].GetYaxis().SetLabelFont(132)
    cN[particle] = ROOT.TCanvas("n%ss" % particle, "Number of %s" % particle, 800, 600)
    cN[particle].SetBottomMargin(0.15)
    cN[particle].SetLeftMargin(0.2)
    hN[particle].Draw()
    cN[particle].Draw()
    filename = 'n%ss.png' % particle
    cN[particle].SaveAs( filename )
```
