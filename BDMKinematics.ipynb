```python
import ROOT
import math
ROOT.gStyle.SetOptStat(0)
fs = ROOT.TFile('/Users/yuntse/Data/pheno/BoostedDM/dune_v07_06_01/bdm/ana_v2/dune_scalar_e5.5_m5_g1_z1.0_Gen_g4_RecoSmear_ana.root')
```

```python
ts = fs.Get("MCParticles")
```

```python
hsVisibleKin = ROOT.TH2D( 'BDMVisibleKin', 'BDM Visible Kinematics', 50, 0., 1., 20, 0., 1. )
hsRecoKin = ROOT.TH2D( 'BDMRecoKin', 'BDM Reconstruected Kinematics', 50, 0., 1., 20, 0., 1. )
```

```python
for i in ts:
    visibleCosth = math.cos(i.VisibleAngle[0])
    hsVisibleKin.Fill( visibleCosth, i.VisibleP[0] )
    recoCosth = math.cos(i.SmearedReconstructableAngle[0])
    hsRecoKin.Fill( recoCosth, i.SmearedReconstructableP[0] )
```

```python
csVisibleKin = ROOT.TCanvas( 'BDMVisibleKin', "BDM Visible Kinematics", 800, 600 )
csVisibleKin.SetLeftMargin( 0.15 )
csVisibleKin.SetBottomMargin( 0.15 )
hsVisibleKin.GetXaxis().SetLabelSize(0.06)
hsVisibleKin.GetYaxis().SetLabelSize(0.06)
hsVisibleKin.GetXaxis().SetTitle("cos(#theta)")
hsVisibleKin.GetXaxis().SetTitleSize(0.07)
hsVisibleKin.GetYaxis().SetTitle("3-Momentum [GeV]")
hsVisibleKin.GetYaxis().SetTitleSize(0.07)
hsVisibleKin.Draw('COLZ')
csVisibleKin.Draw()
csVisibleKin.SaveAs('BDMVisibleKin.png')
```

```python
csRecoKin = ROOT.TCanvas( 'BDMRecoKin', "BDM Reconstructed Kinematics", 800, 600 )
csRecoKin.SetLeftMargin( 0.15 )
csRecoKin.SetBottomMargin( 0.15 )
hsRecoKin.GetXaxis().SetLabelSize(0.06)
hsRecoKin.GetYaxis().SetLabelSize(0.06)
hsRecoKin.GetXaxis().SetTitle("cos(#theta)")
hsRecoKin.GetXaxis().SetTitleSize(0.07)
hsRecoKin.GetYaxis().SetTitle("3-Momentum [GeV]")
hsRecoKin.GetYaxis().SetTitleSize(0.07)
hsRecoKin.Draw('COLZ')
csRecoKin.Draw()
csRecoKin.SaveAs('BDMRecoKin.png')
```
