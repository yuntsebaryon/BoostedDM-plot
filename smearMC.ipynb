```python
import ROOT
import math
ROOT.gStyle.SetOptStat(0)
f = ROOT.TFile("/Users/yuntse/Work/pheno/BoostedDM/gallery/srcs/SmearedMCParticles.root", "READ")
```

```python
t = f.Get("MCParticles")
```

```python
hnProtons = ROOT.TH1F("nProtons", "Number of Protons", 10, 0, 10)
hnParticles = ROOT.TH1F("nParticles", "Number of Particles", 10, 0, 10)
hLeadingProtonPx = ROOT.TH1D("LeadingProtonPx", "Leading Proton Px", 100, -0.5, 0.5)
hLeadingParticleCosAngle = ROOT.TH1D("LeadingParticleCosAngle", "Leading Particle Cos(Angle)", 50, -1., 1. )
hLeadingParticleNoNCosAngle = ROOT.TH1D("LeadingParticeNoNCosAngle", "Leading Particle (without neutrons) Cos(Angle)", 50, -1., 1. )
hVisibleCosAngle = ROOT.TH1D("VisibleCosAngle", "Visible Cos(Angle)", 50, -1., 1. )
hVisibleNoNCosAngle = ROOT.TH1D("VisibleNoNCosAngle", "Visible (without neutrons) Cos(Angle)", 50, -1., 1. )
hInDMP = ROOT.TH1D("InDMP", "Incoming DM Momentum", 55, 12.5, 18. )
hOutDMP = ROOT.TH1D("OutDMP", "Outgoing DM Momentum", 55, 12.5, 18. )
hOutDMAngle = ROOT.TH1D("OutDMAngle", "Outgoing DM Cos(Angle)", 50, -0.2, 0.2 )
hSmearedVisibleCosAngle = ROOT.TH1D("SmearedVisibleCosAngle", "Visible Cos(Angle) after Smearing", 50, -1., 1. )
hSmearedVisibleNoNCosAngle = ROOT.TH1D("SmearedVisibleNoNCosAngle", "Visible (without neutrons) Cos(Angle) after Smearing", 50, -1., 1. )
hSmearedRecontableCosAngle = ROOT.TH1D("SmearedReconstructableCosAngle", "Reconstructable Cos(Angle) after smearing", 50, -1., 1.)
hSmearedRecontableNoNCosAngle = ROOT.TH1D("SmearedReconstructableNoNCosAngle", "Reconstructable (without neutrons) Cos(Angle) after smearing", 50, -1., 1.)
hLeadingSmearedCosAngle = ROOT.TH1D("LeadingSmeared", "Cos(Angle) of the leading particle after smearing", 50, -1., 1.)
hLeadingSmearedNoNCosAngle = ROOT.TH1D("LeadingSmearedNoN", "Cos(Angle) of the leading particle (without neutron) after smearing", 50, -1., 1. )
hLeadingRecontableCosAngle = ROOT.TH1D("LeadingRecontructableCosAngle", "Cos(Angle) of the leading recontructable particle after smearing", 50, -1., 1. )
hLeadingRecontableNoNCosAngle = ROOT.TH1D("LeadingReconstructableNoNCosAngle", "Cos(Angle) of the leading reconstructable particle (without neutrons) after smearing", 50, -1., 1. )
```

```python
nLeadingParticleCosAngle = 0
nLeadingParticleNoNCosAngle = 0
nVisibleCosAngle = 0
nVisibleNoNCosAngle = 0
nSmearedVisibleCosAngle = 0
nSmearedVisibleNoNCosAngle = 0
nRecontableCosAngle = 0
nRecontableNoNCosAngle = 0
nLeadingSmearedCosAngle = 0
nLeadingSmearedNoNCosAngle = 0
nLeadingRecontableCosAngle = 0
nLeadingRecontableNoNCosAngle = 0

for i in t:
    hnProtons.Fill(i.nProtons)
    if len(i.ProtonPx) > 0:
        hLeadingProtonPx.Fill(i.ProtonPx[0])
    LeadingParticleCosAngle = math.cos(i.LeadingParticleAngle[0])
    hLeadingParticleCosAngle.Fill(LeadingParticleCosAngle)
    if LeadingParticleCosAngle > 0.7:
        nLeadingParticleCosAngle += 1
    LeadingParticleNoNCosAngle = math.cos(i.LeadingParticleNoNAngle[0])
    hLeadingParticleNoNCosAngle.Fill(LeadingParticleNoNCosAngle)
    if LeadingParticleNoNCosAngle > 0.7:
        nLeadingParticleNoNCosAngle += 1
    VisibleCosAngle = math.cos(i.VisibleAngle[0])
    hVisibleCosAngle.Fill(VisibleCosAngle)
    if VisibleCosAngle > 0.7:
        nVisibleCosAngle += 1
    VisibleNoNCosAngle = math.cos(i.VisibleNoNAngle[0])
    hVisibleNoNCosAngle.Fill(VisibleNoNCosAngle)
    if VisibleNoNCosAngle > 0.7:
        nVisibleNoNCosAngle += 1
    hInDMP.Fill(i.InDMP[0])
    hOutDMP.Fill(i.OutDMP[0])
    hOutDMAngle.Fill(i.OutDMAngle[0])
    SmearedVisibleCosAngle = math.cos(i.SmearedVisibleAngle[0])
    hSmearedVisibleCosAngle.Fill(SmearedVisibleCosAngle)
    if SmearedVisibleCosAngle > 0.7:
        nSmearedVisibleCosAngle += 1
    SmearedVisibleNoNCosAngle = math.cos(i.SmearedVisibleNonAngle[0])
    hSmearedVisibleNoNCosAngle.Fill(SmearedVisibleNoNCosAngle)
    if SmearedVisibleNoNCosAngle > 0.7:
        nSmearedVisibleNoNCosAngle += 1
    SmearedRecontableCosAngle = math.cos(i.SmearedReconstructableAngle[0])
    hSmearedRecontableCosAngle.Fill(SmearedRecontableCosAngle)
    if SmearedRecontableCosAngle > 0.7:
        nRecontableCosAngle += 1
    SmearedRecontableNoNCosAngle = math.cos(i.SmearedReconstructableNoNAngle[0])
    hSmearedRecontableNoNCosAngle.Fill(SmearedRecontableNoNCosAngle)
    if SmearedRecontableNoNCosAngle > 0.7:
        nRecontableNoNCosAngle += 1
    LeadingSmearedCosAngle = math.cos(i.LeadingSmearedAngle[0])
    hLeadingSmearedCosAngle.Fill(LeadingSmearedCosAngle)
    if LeadingSmearedCosAngle > 0.7:
        nLeadingSmearedCosAngle += 1
    LeadingSmearedNoNCosAngle = math.cos(i.LeadingSmearedNoNAngle[0])
    hLeadingSmearedNoNCosAngle.Fill(LeadingSmearedNoNCosAngle)
    if LeadingSmearedNoNCosAngle > 0.7:
        nLeadingSmearedNoNCosAngle += 1
    LeadingRecontableCosAngle = math.cos(i.LeadingSmearedReconstructableAngle[0])
    hLeadingRecontableCosAngle.Fill(LeadingRecontableCosAngle)
    if LeadingRecontableCosAngle > 0.7:
        nLeadingRecontableCosAngle += 1
    LeadingRecontableNoNCosAngle = math.cos(i.LeadingSmearedReconstructableNoNAngle[0])
    hLeadingRecontableNoNCosAngle.Fill(LeadingRecontableNoNCosAngle)
    if LeadingRecontableNoNCosAngle > 0.7:
        nLeadingRecontableNoNCosAngle += 1
```

```python
cnProtons = ROOT.TCanvas("nProtons", "Number of Protons Generated", 800, 600)
hnProtons.Draw()
cnProtons.Draw()
```

```python
cLeadingParticleCosAngle = ROOT.TCanvas("LeadingParticleCosAngle", "Leading Particle Cos(Angle)", 800, 600)
cLeadingParticleCosAngle.SetBottomMargin(0.15)
hLeadingParticleCosAngle.SetLineColor(593)
hLeadingParticleCosAngle.SetLineWidth(3)
hLeadingParticleCosAngle.SetTitle("")
hLeadingParticleCosAngle.GetXaxis().SetTitle("cos(#theta)")
hLeadingParticleCosAngle.GetXaxis().SetTitleSize(0.07)
hLeadingParticleCosAngle.GetXaxis().SetLabelSize(0.06)
hLeadingParticleCosAngle.GetYaxis().SetLabelSize(0.06)
hLeadingParticleCosAngle.Draw()
hLeadingParticleNoNCosAngle.SetLineColor(906)
hLeadingParticleNoNCosAngle.SetLineWidth(3)
hLeadingParticleNoNCosAngle.Draw("same")
lLeadingParticleCosAngle = ROOT.TLegend( 0.15, 0.6, 0.55, 0.8 )
lLeadingParticleCosAngle.SetBorderSize(0)
lLeadingParticleCosAngle.SetFillStyle(0)
lLeadingParticleCosAngle.AddEntry( hLeadingParticleCosAngle, "With Neutrons")
lLeadingParticleCosAngle.AddEntry( hLeadingParticleNoNCosAngle, "No Neutron")
lLeadingParticleCosAngle.Draw()
cLeadingParticleCosAngle.Draw()
cLeadingParticleCosAngle.SaveAs('LeadingParticleCosAngle.png')
```
