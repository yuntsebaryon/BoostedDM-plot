```python
import ROOT
ROOT.gStyle.SetOptStat(0)
f = ROOT.TFile("/Users/yuntse/Work/pheno/BoostedDM/gallery/srcs/MCParticleDistributions.root", "READ")
```

```python
t = f.Get("MCParticles")
```

```python
h1 = ROOT.TH1F("nProton", "Number of Protons", 10, 0, 10)
h2 = ROOT.TH1D("LeadingProtonPx", "Leading Proton Px", 100, -0.5, 0.5)
```

```python
for i in t:
    h1.Fill(i.nProtons)
    if len(i.ProtonPx) > 0:
        h2.Fill(i.ProtonPx[0])
```

```python
c1 = ROOT.TCanvas("nProtons", "Number of Protons Generated", 800, 600)
h1.Draw()
c1.Draw()
```

```python
c2 = ROOT.TCanvas("LeadingProtonPx", "Leading Proton Px", 800, 600)
h2.Draw()
c2.Draw()
```
